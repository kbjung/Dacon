# 종목 별 주식 종가 수집

```python
import FinanceDataReader as fdr
from tqdm import tqdm
import pandas as pd
```

### 종목코드 파일 불러오기

```python
stock_list = pd.read_csv('../data/stock_list.csv')
```

```python
stock_list['종목코드'] = stock_list['종목코드'].astype(str).str.zfill(6)
```

```python
stock_list.head()
```

## 수집할 기간 설정

### 1주 종가

```python
start_date = '2021-11-01'
end_date = '2021-11-05'
```

```python
Business_days = pd.DataFrame(pd.date_range(start=start_date, end=end_date, freq='B'), columns=['Date'])
Business_days
```

```python
stock_end_list = {}
for code in tqdm(stock_list['종목코드'].values):
    data = fdr.DataReader(code, start = start_date, end = end_date)[['Close']].reset_index()
    stock_end_list[code] = data['Close']
```

```python
len(stock_end_list)
```

```python
Business_days['Date'] = Business_days['Date'].astype(str)
index = list(Business_days['Date'].values)
index
```

```python
end_list = pd.DataFrame(stock_end_list)
end_list
```

```python
end_list.index = index
```

```python
end_list
```

```python
end_list.to_csv('20211101_05_end.csv', index=True)
```
