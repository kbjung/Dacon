# # 매칭 별 모델 학습/예측

```python
# 순서형 특성 수치형으로 지정 : ⭕
    # 'person_attribute_a_a', 'person_attribute_b', 'person_prefer_e', 'contents_attribute_e'
```

```python
import itertools
```

```python
# 단계(4단계)별로 [0, 1]의 선택지 있는 모든 경우의 수 출력
match_cases = list(itertools.product(*[[0,1] for i in range(2)]))
match_cases
```

```python
num_list = [0, 1, 2, 3]
```

```python
pred = pd.DataFrame()
```

## 1번째

```python
num = num_list[0]
c0, c1 = match_cases[num]

tmp_train = train[(train['d_l_match_yn']==c0)&(train['h_l_match_yn']==c1)]
tmp_test = test[(test['d_l_match_yn']==c0)&(test['h_l_match_yn']==c1)]

ex = setup(tmp_train, target='target', n_jobs=-1, session_id=0, verbose=False, silent=True, 
  fold=5, fold_shuffle=True, combine_rare_levels=True,
  numeric_features=['person_attribute_a_a', 'person_attribute_b', 'person_prefer_e', 'contents_attribute_e'],
  normalize=True, normalize_method='minmax'
)
```

```python
compare_models(sort='F1')
# 2m 32s
```

```python
# 모델 생성
model1 = create_model('nb')
model2 = create_model('catboost')

# 블랜드
blended_model = blend_models([model1, model2], fold=5, optimize='F1', method='soft')
```

```python
# 예측
tmp_pred = predict_model(blended_model, tmp_test)
pred = pd.concat([pred, tmp_pred])
```
